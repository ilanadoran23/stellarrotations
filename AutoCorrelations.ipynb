```python
#McQuillan et al Autocorrelation Function with the number of points fixed at 100

def ac_mcquillan_set_k(star): 
    varied_k = [] #array of autocorrelated fluxes for a given star with varied lag times
    flux = star.flux[0:] 
    N = len(flux) 
    time = star.time[0:]
    dt = time[1] - time[0] #cadence
    k = np.linspace(5, N/2, 100, dtype = int) #lag time
    flux_bar = np.average(flux)
    
    #equation for McQuillan's AC function
    for i, v in enumerate(k): 
        numerator = []
        for ii in range(N - k[i]): 
            num = (flux[ii]-flux_bar)*(flux[ii + k[i]] - flux_bar)  
            numerator.append(num)
            
        denominator = []
        for jj in range(N):
            den = (flux[jj]-flux_bar)**2
            denominator.append(den)
            
        rk = (np.sum(numerator)/np.sum(denominator))
        varied_k.append(rk)
    
    tau = k * dt 

    return varied_k, tau 
```

```python
#McQuillan function with adjustable resolution 
def ac_mcquillan(star, num_points): 
    varied_k = [] #array of autocorrelated fluxes for a given star with varied lag time 
    flux = star.flux[0:] 
    N = len(flux)
    time = star.time[0:]
    dt = time[1] - time[0] #cadence
    k = np.linspace(5, (2*N/3), num_points, dtype = int) #lag time
    flux_bar = np.average(flux)
    
    
    #equation for McQuillan's AC function
    for i, v in enumerate(k): 
        numerator = []
        for ii in range(N - k[i]): 
            num = (flux[ii]-flux_bar)*(flux[ii + k[i]] - flux_bar)  
            numerator.append(num)
            
        denominator = []
        for jj in range(N):
            den = (flux[jj]-flux_bar)**2
            denominator.append(den)
            
        rk = (np.sum(numerator)/np.sum(denominator))
        varied_k.append(rk)
    
    tau = k * dt

    return varied_k, tau 
```

# Errors in Periods

#McQuillan : σ = (1.483 × MAD)/ √N-1

def period_error(star, num_points):
    varied_k = [] #array of autocorrelated fluxes for a given star with varied lag  
    flux = star.flux[0:] 
    N = len(flux)
    time = star.time[0:]
    dt = time[1] - time[0] #cadence
    k = np.linspace(5, N/2, num_points, dtype = int) #lag 
    flux_bar = np.average(flux)
    
    for i, v in enumerate(k): 
        numerator = []
        for ii in range(N - k[i]): 
            num = (flux[ii]-flux_bar)*(flux[ii + k[i]] - flux_bar)  
            numerator.append(num)
            
        denominator = []
        for jj in range(N):
            den = (flux[jj]-flux_bar)**2
            denominator.append(den)
            
        rk = (np.sum(numerator)/np.sum(denominator))
        varied_k.append(rk)
    
    tau = k * dt         
    mad_tau = astat.median_absolute_deviation(tau) 
    
    error = (1.483 * mad_tau)/(np.sqrt(N - 1))
    
    return error

# Finding Peaks 

```python
def find_peaks(y): #finding peak positions in a given array
    peaks = []
    for index, value in enumerate(y):
        if value > y[np.int(index - 1)] or value == y[np.int(index - 1)]:
            if index == len(y)-1:
                break
            if value > y[np.int(index + 1)] or value == y[np.int(index + 1)]:
                peaks.append(value)  
    return(peaks)
```

```python
#function for obtaining starspot periods
def ss_ac(Kepler_ID):
    tess_cadence = 1./24./30.
    ss_ac_tpf = search_targetpixelfile('KIC ' + Kepler_ID).download(quality_bitmask='default');
    lc_ss_ac = ss_ac_tpf.to_lightcurve(aperture_mask=ss_ac_tpf.pipeline_mask).remove_outliers(sigma=5).bin(binsize=10);

    med_ss_ac = np.median(lc_ss_ac.flux)
    time_ss_ac = lc_ss_ac.time[0:]
    flux_ss_ac = lc_ss_ac.flux[0:]/med_ss_ac - 1
    yerr_ss_ac = lc_ss_ac.flux_err[0:]/med_ss_ac

    ss_ac_mask = astat.sigma_clip(flux_ss_ac, sigma=1, masked = True)
    flux_ss_ac = flux_ss_ac[~ss_ac_mask.mask]
    time_ss_ac = time_ss_ac[~ss_ac_mask.mask]
    flux_err_ss_ac = yerr_ss_ac[~ss_ac_mask.mask]

    rotate_ss_ac = ss.RotationModel(time_ss_ac, flux_ss_ac, flux_err_ss_ac)

    return rotate_ss_ac.acf_rotation(tess_cadence)
```
